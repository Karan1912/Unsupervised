Normalization

```python
import numpy as np
a = np.random.rand(10)*10
print(a)
```

```python
norm = np.linalg.norm(a)
```

```python
normal_array = a/norm
print(normal_array)
```

```python
from sklearn import preprocessing
b = np.array([4,5,7,4,3,2,6,8,9])
normalized = preprocessing.normalize([b])
print(normalized)
```

Euclidean Distance

```python
a = np.array([1,3,4,7,8])
b = np.array([3,5,6,8,9])
```

```python
temp = a-b
sum_sq = np.dot(temp.T,temp)
```

```python
print(np.sqrt(sum_sq))
```
